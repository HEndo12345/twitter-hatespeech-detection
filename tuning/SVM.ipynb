```python
import optuna
```

```python
def objective(trial):
    params = {
    'norm': trial.suggest_categorical('ngram',['l2', 'l1']),
    'C': trial.suggest_uniform('C', 0.1, 1),
    'loss': trial.suggest_categorical('loss', ['squared_hinge', 'hinge'])
    }
    sw = stopwords.words('english') + ['hes']
    data = list(zip(train_X, train_Y))
    random.shuffle(data)
    ff_train, ff_train_a = zip(*data)
    ff_train = [' '.join(text) for text in ff_train]
    vect = TfidfVectorizer(stop_words=sw,tokenizer=tokenize, 
                          min_df= 0.001, max_df=0.8, 
                          ngram_range=(1,2),
                          norm = params['norm']
                          )
    tfidf = vect.fit_transform(ff_train)
    lsvc = LinearSVC(
            C = params['C'],
            loss = params['loss'],
            max_iter = 3000
    )


    scores = cvt(lsvc, tfidf, ff_train_a, cv = 4, return_train_score= True, scoring = 'f1_macro', )
    test = scores['test_score'].mean()
    train = scores['train_score'].mean()
    f_score = 1.0 - float(test) + float(abs(train-test))/2.0
    return f_score
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
study = optuna.create_study()
study.optimize(objective, n_trials=1000)
```
